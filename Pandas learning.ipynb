### Data Manipulation

```python
data.drop(columns = ['Gender'])
```

```python
data.sort_values(by ='LoanAmount')
```

```python
def customfunc(ApplicantIncome):
    return ApplicantIncome.max()-ApplicantIncome.min()
a=data.groupby('Gender').agg({'LoanAmount' : 'min'})
b=data.groupby('Gender').agg({'LoanAmount' : 'max'})
c=data.groupby('Gender').agg({'LoanAmount' : 'mean'})
d=data.groupby('Gender').agg({'LoanAmount' : 'median'})
e=data.groupby('Gender').agg({'LoanAmount' : 'std'})
f=data.groupby('Gender').agg({'LoanAmount' : 'var'})
g=data.groupby('Gender').agg({'LoanAmount' : 'sum'})
h=data.groupby('Gender').agg({'LoanAmount' : 'count'})
i=data.groupby('Gender').agg({'LoanAmount' : customfunc})
```
